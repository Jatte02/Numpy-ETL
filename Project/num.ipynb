```python
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv'

# Pandas descarga y lee autom√°ticamente
df = pd.read_csv(url)

# Convertimos a numpy array 
# Nota: Pandas ya detecta los encabezados, no necesitas skiprows ni usecols manual
datos = df.iloc[:, 1:6].to_numpy() 

print(datos)
```

```python
diametro_naranja = datos[:5000, 0]
peso_naranja = datos[:5000, 1]

diametro_toronja = datos[5000:, 0]
peso_toronja = datos[5000:, 1]

import matplotlib.pyplot as plt

plt.plot(diametro_naranja, peso_naranja)
plt.plot(diametro_toronja, peso_toronja)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una secuencia de valores de x de -1 a 1
x = np.arange(-1, 1, 0.0001)

# Implementaci√≥n de la f√≥rmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Graficar el gr√°fico con las dos partes del c√≠rculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )

# Agregar el t√≠tulo del gr√°fico y las etiquetas de los ejes x e y
plt.title("C√≠rculo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

# Mostrar el gr√°fico
plt.show()
```
