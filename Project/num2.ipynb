```python
import pandas as pd
import requests
url = 'https://es.wikipedia.org/wiki/Anexo:Pa%C3%ADses_y_territorios_dependientes_por_poblaci%C3%B3n'

# 1. Definimos un "User-Agent" (esto imita a un navegador Chrome en Windows)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

# 2. Hacemos la petición usando requests con los headers
respuesta = requests.get(url, headers=headers)

# 3. Le pasamos el contenido de la respuesta (.text) a pandas
# Nota: read_html devuelve una lista de tablas, la [0] suele ser la principal
df = pd.read_html(respuesta.text)[0]

# Mostramos los primeros resultados
print(df.head())

```
