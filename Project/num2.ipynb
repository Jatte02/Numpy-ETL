# Proyecto 1

```python
import pandas as pd

url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
datos = pd.read_json(url, storage_options={'User-Agent': 'Mozilla/5.0'})
print(datos.head())
```

```python
datos = pd.json_normalize(datos['dados_vendas'])
print(datos.head())
```

```python
columnas = list(datos.columns)
print(columnas)
```

```python
datos = datos.explode(columnas[1:])
print(datos.head())
```

```python
datos.reset_index(drop=True, inplace=True)
print(datos.head())
```

```python
import numpy as np

datos['Valor da compra'] = datos['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',','.').strip())

datos['Valor da compra'] = datos['Valor da compra'].astype(np.float64)
```

# Proyecto 2

```python
url2 = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'

datos2 = pd.read_json(url2, storage_options={'User-Agent': 'Mozilla/5.0'})
print(datos2.head())
```

```python
datos2 = pd.json_normalize(datos2['dados_locacao'])
```

```python
print(datos2.head())
```

```python
columnas2 = list(datos2.columns)
columnas2
```

```python
datos2 = datos2.explode(columnas2[1:])
datos2.reset_index(drop=True, inplace=True)

datos2
```

```python
datos2['valor_aluguel'] = datos2['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',','.').replace('reais','').strip())
datos2['valor_aluguel'] = datos2['valor_aluguel'].astype(np.float64)
```
